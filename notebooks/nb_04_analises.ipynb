**Objetivo:**
Responder às perguntas de negócio definidas no objetivo do MVP,
utilizando o modelo analítico construído na camada Gold, além de
avaliar a qualidade dos dados utilizados nas análises.



```python
%sql
USE CATALOG mvp_catalog;
USE SCHEMA gold;

```

```python
%sql
SELECT COUNT(*) AS total_linhas
FROM fact_sales;



```

```python
%sql
SELECT
  SUM(line_total) AS receita_total
FROM fact_sales;
```

```python
%sql
SELECT
  t.territory_name,
  SUM(f.line_total) AS receita_total
FROM fact_sales f
JOIN dim_territory t
  ON f.territory_id = t.territory_id
GROUP BY t.territory_name
ORDER BY receita_total DESC;
```

```python
%sql
SELECT
  t.territory_name,
  AVG(f.line_total) AS ticket_medio
FROM fact_sales f
JOIN dim_territory t
  ON f.territory_id = t.territory_id
GROUP BY t.territory_name
ORDER BY ticket_medio DESC;
```

```python
%sql
SELECT
  d.year,
  d.month,
  d.month_name,
  SUM(f.line_total) AS receita
FROM fact_sales f
JOIN dim_date d
  ON f.order_date = d.date
GROUP BY d.year, d.month, d.month_name
ORDER BY d.year, d.month;
```

```python
%sql
SELECT
  p.product_name,
  SUM(f.line_total) AS faturamento
FROM fact_sales f
JOIN dim_product p
  ON f.product_id = p.product_id
GROUP BY p.product_name
ORDER BY faturamento DESC
LIMIT 10;
```

```python
%sql
SELECT
  p.product_name,
  SUM(f.order_qty) AS quantidade_vendida
FROM fact_sales f
JOIN dim_product p
  ON f.product_id = p.product_id
GROUP BY p.product_name
ORDER BY quantidade_vendida DESC
LIMIT 10;
```

```python
%sql
SELECT
  p.product_name,
  d.month_name,
  SUM(f.line_total) AS receita
FROM fact_sales f
JOIN dim_product p
  ON f.product_id = p.product_id
JOIN dim_date d
  ON f.order_date = d.date
GROUP BY p.product_name, d.month_name
ORDER BY p.product_name, receita DESC;
```

```python
%sql
SELECT
  c.customer_id,
  c.first_name,
  c.last_name,
  COUNT(DISTINCT f.sales_order_id) AS total_pedidos
FROM fact_sales f
JOIN dim_customer c
  ON f.customer_id = c.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name
ORDER BY total_pedidos DESC;
```

```python
%sql
SELECT
  c.customer_id,
  c.first_name,
  c.last_name,
  SUM(f.line_total) AS valor_total
FROM fact_sales f
JOIN dim_customer c
  ON f.customer_id = c.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name
ORDER BY valor_total DESC;
```

```python
%sql
SELECT
  COUNT(*) AS linhas_fato,
  COUNT(DISTINCT customer_id) AS clientes_distintos,
  COUNT(DISTINCT product_id) AS produtos_distintos,
  COUNT(DISTINCT territory_id) AS territorios_distintos
FROM fact_sales;
```
